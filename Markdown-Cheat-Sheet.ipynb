# Markdown Cheat Sheet


## Paragraphs

```
This is a paragraph

This is a second paragraph.
```

This is a paragraph

This is a second paragraph.

## Headers

```
# H1
## H2
### H3
#### H4
##### H5
###### H6
```

## H2
### H3
#### H4
##### H5
###### H6

## Emphasis

```
This is **BOLD**
This is *italic*
```
This is **BOLD**<br>
This is *italic*

## Unorderd list

```
- item 1
- item 2
- item 3
```

- item 1
- item 2
- item 3

## Ordered list

```
1. item 1
2. item 2
3. item 3
```

1. item 1
2. item 2
3. item 3
